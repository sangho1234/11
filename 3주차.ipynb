조건문

```python
temp = 8   

if temp >= 20:
    print("덥다")
elif temp > 10:
    print("따듯함") 
elif temp >= 0:
    print("춥다") 
else:
    print("매우 춥다")
```

반복문과 조건문 함께 사용

```python
scores = [42,60,77,89,20]

for score in scores:
    if score >= 60:
        print(score,"합격")
    else:
         print(score,"불합격")
         
```
