```python
# 할인율 계산
# discount: 0.0~1.0(0%~100%)
def calculate_price(price, discount=0):
    discounted = price * (1 - discount)
    return discounted

print(f"기본 가격: {calculate_price(10000)}원")
print(f"할인 가격: {calculate_price(10000, 0.1)}원")
```

함수 여러값 계산하기

```python
#최솟값과 최댓값을 동시 반환하는 함순
def min_max(numbers):
    minimum = min(numbers)
    maxium = max(numbers)
    return minimum,maxium
scores = [10,20,30,40,50,60,70,80]
low, high = min_max(scores)
print(min_max(scores))
print("최댓값:", high)
```

```python
# 학생 데이터
grade_data = {

"name": ["kim","lee","park","cho","choi","jung"],
"korean": [88,72,85,65,80,91],
"math": [88,72,85,65,80,91],
"english": [88,72,85,65,80,91],

}
grade_df = pd.DataFrame(grade_data)

print(grade_df)
```

정렬하기

```python
print("---국어 점수 낮은 순----")
print(grade_df.sort_values("korean"))
print()
print("-----수학 점수 높은 순------")
print(grade_df.sort_values("math",ascending=False))
```

## 새로운 열 추가하기

```python

grade_df["average"] = (grade_df["korean"] + grade_df["math"] + grade_df["english"]) / 3
grade_df["average"] = grade_df["average"].round(1)
grade_df
```

```python
grade_df["grade"] = grade_df["average"].apply(
    lambda x: "A" if x >= 80 else "B"
)
grade_df
```

그룹별 통계

```python
group_avg = grade_df.groupby("grade")["average"].mean()
print("등급별 평균:")
print(group_avg)
```

```python
# 응용 - 회사 부서별 급여 파악

employee_data = {

    "name": ["kim","lee","park","cho","choi","jung"],
    "department":["개발","마케팅","개발","인사","마케팅","인사"],
    "salary":[4200,3500,4800,3200,3800,5000]
}

emp_df = pd.DataFrame(employee_data)
# 부서별 평균 
dept_avg = emp_df.groupby("department")["salary"].mean()
print(dept_avg.round(0))
print()
print("----부서별 인원 수-----")
dept_count = emp_df.groupby("department")["name"].count()
print(dept_count)

```
