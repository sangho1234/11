```python
def analyze_scores(scores):
    # 평균 계산 (전체 합계를 개수로 나눔)
    avg = sum(scores) / len(scores)
    
    # 최고점과 최저점 (파일의 min, max 활용)
    high = max(scores)
    low = min(scores)
    
    # 합격자 수 (반복문과 조건문 활용)
    pass_count = 0
    for s in scores:
        if s >= 60:
            pass_count += 1
            
    # 출력 (파일의 f-string 포맷 활용)
    print(f"- 평균: {avg:.1f}점")
    print(f"- 최고점: {high}점")
    print(f"- 최저점: {low}점")
    print(f"- 합격(60점 이상): {pass_count}명")

# 문제 데이터
class_scores = [45, 62, 78, 55, 90, 83, 71, 66, 48, 95]
analyze_scores(class_scores)
```
